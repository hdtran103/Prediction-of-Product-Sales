Are there duplicates? If so, drop any duplicates

```python
# Identify duplicate rows in the DataFrame:
duplicate_rows = sp[sp.duplicated()]
duplicate_rows
```

```python
# Remove duplicate rows from the DataFrame:
sp = sp.drop_duplicates()
sp.drop_duplicates()
```

- - Identify Missing Values.

```python
# Checking for missing values:
missing_values = sp.isnull()
missing_values
```

Total of Missing Values.

```python
# print the number of missing values:
num_missing = sp.isnull().sum()
num_missing
```

- Decide on how to address the missing values 
- using the fillna() method, to fill all missing values in the 'Outlet_Size' column with the string'unknown' 



```python
# Fill all missing values with a specific values:
sp['Outlet_Size'] = sp['Outlet_Size'].fillna('unknown')
item_weight_median = sp['Item_Weight'].median()
sp['Item_Weight'] = sp['Item_Weight'].fillna(item_weight_median)
```

- Confirmation of no missing values

```python
# print summary information for confirmation of missing values:
sp.info()
```

-  Find and fix any inconsistent categories of data.

```python
# Fix inconsistent categories of data:
sp['Item_Fat_Content'].value_counts()
sp['Item_Fat_Content'] = sp['Item_Fat_Content'].replace({'low fat': 'Low Fat', 'LF': 'Low Fat', 'reg': 'Regular'})
```
