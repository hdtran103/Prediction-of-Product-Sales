- Fills in missing values in the 'Item_Weight' column by first grouping the dataset by 'Item_Identifier' 
- Calculating the mean weight for each item. 
- It will identifies which rows have missing values for 'Item_Weight' and uses the apply() function to replace those missing values with the mean weight for the corresponding item.

```python
# Fill in missing values in 'Item_Weight' column.
item_weight_means = sp.groupby('Item_Identifier')['Item_Weight'].mean()
missing_weights = sp['Item_Weight'].isnull()
sp.loc[missing_weights, 'Item_Weight'] = sp.loc[missing_weights, 'Item_Identifier'].apply(lambda x: item_weight_means[x])
```

- Fills in missing values in the 'Outlet_Size' column using the mode of the column. 
- The mode is the value that appears most frequently in the column. 
- First, the mode is calculated using the .mode() method, which returns a Pandas series with the mode(s) of the column. 
- Take the first value of this series using [0]. 
- Next, we create a boolean mask called missing_sizes to identify the rows where the 'Outlet_Size' column has missing values. 
- Use the .loc[] method to fill in the missing values with the mode of the column.

```python
# Fill in missing values in 'Outlet_Size' column.
outlet_size_mode = sp['Outlet_Size'].mode()[0]
missing_sizes = sp['Outlet_Size'].isnull()
sp.loc[missing_sizes, 'Outlet_Size'] = outlet_size_mode
```

- This ensures that the column contains consistent categories for 'Low Fat' and 'Regular'
- The value_counts() method is used to verify that the inconsistent categories have been replaced.
- This code replaces inconsistent categories in the 'Item_Fat_Content' column

```python
# Replace inconsistent categories in 'Item_Fat_Content' column.
sp['Item_Fat_Content'].value_counts()
sp['Item_Fat_Content'] = sp['Item_Fat_Content'].replace({'low fat': 'Low Fat', 'LF': 'Low Fat', 'reg': 'Regular'})
```

# Item Outlet Sales 
- The code is plotting a histogram of the 'Item_Outlet_Sales'in the dataset to visualize the distribution of the variable. 
- This is done to gain insights into the shape of the distribution, any outliers or skewness, and the overall range of values in the dataset, which can be useful for further analysis and modeling. 
- The histogram shows the frequency distribution of the 'Item_Outlet_Sales' variable, with the x-axis showing the range of sales
- The y-axis showing the frequency of sales within that range. 
- The 'kde=True' argument adds a kernel density estimate to the histogram, which can provide a smooth estimate of the distribution's shape.
